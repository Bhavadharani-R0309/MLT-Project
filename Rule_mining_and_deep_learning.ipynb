```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
df['average_score'] = (
    df['math score'] +
    df['reading score'] +
    df['writing score']
) / 3

df['dropout'] = df['average_score'].apply(lambda x: 1 if x < 40 else 0)

df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop('dropout', axis=1)
y = df['dropout']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Build model
model = Sequential()
model.add(Dense(16, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=20, batch_size=10)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Create binary dataset
df_arm = df.copy()

df_arm['math_low'] = df_arm['math score'] < 40
df_arm['reading_low'] = df_arm['reading score'] < 40
df_arm['writing_low'] = df_arm['writing score'] < 40

df_arm = df_arm[['math_low', 'reading_low', 'writing_low', 'dropout']]
df_arm = df_arm.astype(bool)

# Frequent itemsets
frequent_itemsets = apriori(df_arm, min_support=0.02, use_colnames=True)

print("\n==== FREQUENT ITEMSETS ====\n")
print(frequent_itemsets.head())

# Rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print("\n==== ASSOCIATION RULES ====\n")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

```python
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score

# Naive Bayes
start = time.time()
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
nb_time = time.time() - start

nb_acc = accuracy_score(y_test, y_pred_nb)
nb_f1 = f1_score(y_test, y_pred_nb)

# ANN
start = time.time()
y_pred_ann = model.predict(X_test)
y_pred_ann = (y_pred_ann > 0.5)
ann_time = time.time() - start

ann_acc = accuracy_score(y_test, y_pred_ann)
ann_f1 = f1_score(y_test, y_pred_ann)

# Table
import pandas as pd

results = pd.DataFrame({
    "Model": ["Naive Bayes", "ANN"],
    "Accuracy": [nb_acc, ann_acc],
    "F1 Score": [nb_f1, ann_f1],
    "Training Time (s)": [nb_time, ann_time]
})

print("\n===== MODEL COMPARISON =====\n")
print(results)
```
